Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
def diagrama_disp (data, col_num, col_num2, col_cat, col_size, scale=(1,200)):
    fig, axs = plt.subplots(nrows=1,ncols=1, figsize=(10,5))
    args = {
        "data": data,
        "x": col_num,
        "y": col_num2,
        "hue": col_cat,
        "size": col_size,
        "sizes": scale,
        "alpha": 0.5,
        "palette": "viridis",
        "legend": "auto"    
        
    }
    
    return sns.scatterplot(**args);

```

```python
def plot_categorical_relationship_fin(df, cat_col1, cat_col2, relative_freq=False, show_values=False, size_group = 5):
    # Prepara los datos
    count_data = df.groupby([cat_col1, cat_col2]).size().reset_index(name='count')
    total_counts = df[cat_col1].value_counts()
    
    # Convierte a frecuencias relativas si se solicita
    if relative_freq:
        count_data['count'] = count_data.apply(lambda x: x['count'] / total_counts[x[cat_col1]], axis=1)

    # Si hay ms de size_group categoras en cat_col1, las divide en grupos de size_group
    unique_categories = df[cat_col1].unique()
    if len(unique_categories) > size_group:
        num_plots = int(np.ceil(len(unique_categories) / size_group))

        for i in range(num_plots):
            # Selecciona un subconjunto de categoras para cada grfico
            categories_subset = unique_categories[i * size_group:(i + 1) * size_group]
            data_subset = count_data[count_data[cat_col1].isin(categories_subset)]

            # Crea el grfico
            fig = plt.figure(figsize=(10, 6))
            ax = sns.barplot(x=cat_col1, y='count', hue=cat_col2, data=data_subset, order=categories_subset, palette="ocean")

            # Aade ttulos y etiquetas
            #plt.title(f'Relacin entre {cat_col1} y {cat_col2} - Grupo {i + 1}')
            #plt.xlabel(cat_col1)
            #plt.ylabel('Frecuencia' if relative_freq else 'Conteo')
            #plt.xticks(rotation=45)

            # Mostrar valores en el grfico
            if show_values:
                for p in ax.patches:
                    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                                ha='center', va='center', fontsize=10, color='black', xytext=(0, size_group),
                                textcoords='offset points')

            # Muestra el grfico
            plt.show()
            fig.savefig("titanic5.png")
    else:
        # Crea el grfico para menos de size_group categoras
        fig= plt.figure(figsize=(10, 6))
        ax = sns.barplot(x=cat_col1, y='count', hue=cat_col2, data=count_data, palette="ocean")

        # Aade ttulos y etiquetas
        #plt.title(f'Hombres y mujeres muertos vs vivos')
        #plt.xlabel(cat_col1)
        #plt.ylabel('Frecuencia' if relative_freq else 'Conteo')
        #plt.xticks(rotation=45)

        # Mostrar valores en el grfico
        if show_values:
            for p in ax.patches:
                ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                            ha='center', va='center', fontsize=10, color='black', xytext=(0, size_group),
                            textcoords='offset points')

        # Muestra el grfico
        plt.show()
        fig.savefig("titanic5.png")
```

```python
# 3- ... que la clase en la que se viajes tuvo una influencia significativa y podra explicar... 
columnas = ["alive","sex","class"] 
col_directora = "sex"
col_1 = [col for col in columnas if col != col_directora][0]
col_2 = [col for col in columnas if col != col_directora][1]
diccionario_multivariante = {}
for valor in df_titanic[col_directora].unique():
    diccionario_multivariante[valor] = df_titanic.loc[df_titanic[col_directora] == valor,[col_2,col_1]]

for valor,df_datos in diccionario_multivariante.items():
    print(f"{valor}:")
    plot_categorical_relationship_fin(df_datos,col_2,col_1, relative_freq= False, show_values= True)

```

```python
df_titanic4 = df_titanic[["alive","sex","who","class"]]

mujeres=df_titanic4.loc[df_titanic4["who"]=="woman"]
abs_mujeres = mujeres.groupby("class", as_index = False)["alive"].value_counts()
print(abs_mujeres)

nios= df_titanic4.loc[df_titanic4["who"]=="child"]
abs_nios = nios.groupby("class", as_index = False)["alive"].value_counts()
```
